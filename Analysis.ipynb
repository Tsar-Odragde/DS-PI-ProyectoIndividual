# Tabla de compras

```python
compra.describe()
```

Outliers en cantidad

```python
lista_cantidad = sorted(list(compra.Cantidad))
q1 = len(lista_cantidad)//4
q3 = q1 * 3
iqr_cantidad = lista_cantidad[q3] - lista_cantidad[q1]
min_cantidad = 0
max_cantidad = int(lista_cantidad[q3] + (iqr_cantidad * 1.5))
```

```python
outliers_cantidad = [i for i in lista_cantidad if (i > max_cantidad) or (i < min_cantidad)]
len(outliers_cantidad)
```

Outliers en precio

```python
lista_precio = sorted([n for n in list(compra.Precio.dropna())])
q1 = len(lista_precio)//4
q3 = q1 * 3
iqr_precio = lista_precio[q3] - lista_precio[q1]
min_precio = 0
max_precio = lista_precio[q3] + (iqr_precio * 1.5)
```

```python
outliers_precio = [i for i in lista_precio if (i > max_precio) or (i < min_precio)]
len(outliers_precio)
```
